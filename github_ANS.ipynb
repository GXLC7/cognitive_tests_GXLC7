```python
# editing working version of the code from week 5 to send results to google form by adapting "working simple google form example"
from IPython.display import display, HTML, Image, clear_output
import time
import random
random.seed(1)
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
import json
import ipywidgets as widgets
from jupyter_ui_poll import ui_events

def send_to_google_form(data_dict, form_url):
        form_id_start = form_url.find("/e/") + 3  # Find the position after "/e/"
        form_id_end = form_url.find("/", form_id_start)  # Find the next forward slash
        form_id = form_url[form_id_start:form_id_end]
        
        view_form_url = f'https://docs.google.com/forms/d/e/{form_id}/viewform'
        post_form_url = f'https://docs.google.com/forms/d/e/{form_id}/formResponse'
    
        page = requests.get(view_form_url)
    
        content = BeautifulSoup(page.content, "html.parser").find('script', type='text/javascript')
        content = content.text[27:-1]
    
        result = json.loads(content)[1][1]
    
        form_dict = {}
        loaded_all = True
    
        for item in result:
            if item[1] not in data_dict:
                print(f"Form item {item[1]} not found. Data not uploaded.")
                loaded_all = False
                return False
            form_dict[f'entry.{item[4][0][0]}'] = data_dict[item[1]]
    
        post_result = requests.post(post_form_url, data=form_dict)
    
        return post_result.ok

image_list = [Image('p_12_9.png'), Image('p_16_12.png'), Image('p_20_15.png'), Image('p_14_12.png'), Image('p_21_18.png'), 
               Image('p_18_16.png'), Image('p_20_18.png'), Image('p_10_9.png'), Image('g_9_12.png'), Image('g_12_16.png'), 
               Image('g_15_20.png'), Image('g_12_14.png'), Image('g_18_21.png'), Image('g_16_18.png'), Image('g_18_20.png'), 
               Image('g_9_10.png'), Image('p_12_9_a.png'), Image('p_16_12_a.png'), Image('p_20_15_a.png'), Image('p_14_12_a.png'), Image('p_21_18_a.png'), 
               Image('p_18_16_a.png'), Image('p_20_18_a.png'), Image('p_10_9_a.png'), Image('g_9_12_a.png'), Image('g_12_16_a.png'), 
               Image('g_15_20_a.png'), Image('g_12_14_a.png'), Image('g_18_21_a.png'), Image('g_16_18_a.png'), Image('g_18_20_a.png'), 
               Image('g_9_10_a.png'), Image('p_12_9.png'), Image('p_16_12.png'), Image('p_20_15.png'), Image('p_14_12.png'), Image('p_21_18.png'), 
               Image('p_18_16.png'), Image('p_20_18.png'), Image('p_10_9.png'), Image('g_9_12.png'), Image('g_12_16.png'), 
               Image('g_15_20.png'), Image('g_12_14.png'), Image('g_18_21.png'), Image('g_16_18.png'), Image('g_18_20.png'), 
               Image('g_9_10.png'), Image('p_12_9_a.png'), Image('p_16_12_a.png'), Image('p_20_15_a.png'), Image('p_14_12_a.png'), Image('p_21_18_a.png'), 
               Image('p_18_16_a.png'), Image('p_20_18_a.png'), Image('p_10_9_a.png'), Image('g_9_12_a.png'), Image('g_12_16_a.png'), 
               Image('g_15_20_a.png'), Image('g_12_14_a.png'), Image('g_18_21_a.png'), Image('g_16_18_a.png'), Image('g_18_20_a.png'), 
               Image('g_9_10_a.png')
]
#image_list = [Image('g_15_20_a.png'), Image('g_12_14_a.png'), Image('g_18_21_a.png'), Image('g_16_18_a.png'), Image('g_18_20_a.png'), 
#               Image('g_9_10_a.png')
#]

results_dict = {
    'filename': [],
    'nL': [],
    'nR': [],
    'ratio': [],
    'correct_or_incorrect': []
}           
                     
def ANS_test_3():
    data_consent_info = """DATA CONSENT INFORMATION:

    Please read:

    we wish to record your response data
    to an anonymised public data repository. 
    Your data will be used for educational teaching purposes
    practising data analysis and visualisation.

    Please type   yes   in the box below if you consent to the upload."""
    
    print(data_consent_info)
    result = input("> ") 
    
    if result == "yes": 
        print("Thanks for your participation.")
        print("Please contact philip.lewis@ucl.ac.uk")
        print("If you have any questions or concerns")
        print("regarding the stored results.")
        
    else: 
        # end code execution by raising an exception
        raise(Exception("User did not consent to continue test."))
    
    id_instructions = """

    Enter your anonymised ID
    
    To generate an anonymous 4-letter unique user identifier please enter:
    - two letters based on the initials (first and last name) of a childhood friend
    - two letters based on the initials (first and last name) of a favourite actor / actress
    
    e.g. if your friend was called Charlie Brown and film star was Tom Cruise
         then your unique identifer would be CBTC
    """
    
    print(id_instructions)
    user_id = input("> ")
    
    print("User entered id:", user_id)

    print("Please enter your gender:")
    gender = input("")

    print("Welcome to the Approximate Number Sense test")
    time.sleep(1)
         
    print("The following images contain different numbers of pink and green dots. After being presented with each image, state whether there were more pink dots or green dots.")
    time.sleep(3)

    random.shuffle(image_list)
    correct_answers = 0
    
    for image_data in image_list:
        filename = str(image_data.filename)
        #print(filename)
        match = re.search(r'\d+', filename)
        numbers = [int(match) for match in re.findall(r'\d+', filename)]
        #print(type(numbers))
        nL = numbers[0]
        nR = numbers[1]
        if numbers[1] > numbers[0]:
            ratio = numbers[0]/numbers[1]
        elif numbers[0] > numbers [1]:
            ratio = numbers[1]/numbers[0]
        #print(nL, nR)
        #print(ratio)
        
        display(image_data)
        time.sleep(2)
        clear_output(wait=False)

    
        ans = ""
        while ans.lower() not in ["pink", "green"]:
            ans = input("Were there more pink or green dots? Type 'exit' to end the test: ")

            if ans.lower() == 'exit':
                print("Exiting the test.")
                return
        
        answer_first_letter = ans[0].lower()
        correct_answer = image_data.filename[0].lower()

        if answer_first_letter == correct_answer:
            correct_answers += 1
            correct_or_incorrect = 'correct'
        else:
            correct_answers += 0
            correct_or_incorrect = 'incorrect'

        clear_output(wait=True)
             
        results_dict['filename'].append(filename)
        results_dict['nR'].append(nR)
        results_dict['nL'].append(nL)
        results_dict['ratio'].append(ratio)
        results_dict['correct_or_incorrect'].append(correct_or_incorrect)
    
    print("End of test. Total correct answers:", correct_answers)

    results_df = pd.DataFrame(results_dict)
    #print(results_df)

    data_dict = {
        'user_id': user_id,
        'gender': gender,
        'score': correct_answers,
        'results_json': results_df.to_json()
    }

    form_url = 'https://docs.google.com/forms/d/e/1FAIpQLSdvaMm76MB3v2TcSuAV47M9lDvI0EaRZagoVTCzgnEEiuMehw/viewform?usp=sf_link'
    success = send_to_google_form(data_dict, form_url)

    if success:
        print("Data successfully submitted to the Google Form.")
    else:
        print("Failed to submit data to the Google Form.")

ANS_test_3()
```
