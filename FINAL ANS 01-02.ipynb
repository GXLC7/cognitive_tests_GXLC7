```python
from IPython.display import display, HTML, Image, clear_output
import time
import random
random.seed(1)
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
import json
import ipywidgets as widgets
from jupyter_ui_poll import ui_events

def send_to_google_form(data_dict, form_url):
        form_id_start = form_url.find("/e/") + 3  # Find the position after "/e/"
        form_id_end = form_url.find("/", form_id_start)  # Find the next forward slash
        form_id = form_url[form_id_start:form_id_end]
        
        view_form_url = f'https://docs.google.com/forms/d/e/{form_id}/viewform'
        post_form_url = f'https://docs.google.com/forms/d/e/{form_id}/formResponse'
    
        page = requests.get(view_form_url)
    
        content = BeautifulSoup(page.content, "html.parser").find('script', type='text/javascript')
        content = content.text[27:-1]
    
        result = json.loads(content)[1][1]
    
        form_dict = {}
        loaded_all = True
    
        for item in result:
            if item[1] not in data_dict:
                print(f"Form item {item[1]} not found. Data not uploaded.")
                loaded_all = False
                return False
            form_dict[f'entry.{item[4][0][0]}'] = data_dict[item[1]]
    
        post_result = requests.post(post_form_url, data=form_dict)
    
        return post_result.ok

image_list_original = ['p_12_9.png', 
                       'p_16_12.png', 
                       'p_20_15.png', 
                       'p_14_12.png', 
                       'p_21_18.png', 
                       'p_18_16.png', 
                       'p_20_18.png', 
                       'p_10_9.png', 
                       'g_9_12.png', 
                       'g_12_16.png', 
                       'g_15_20.png', 
                       'g_12_14.png', 
                       'g_18_21.png', 
                       'g_16_18.png', 
                       'g_18_20.png', 
                       'g_9_10.png', 
                       'p_12_9_b.png', 
                       'p_16_12_b.png', 
                       'p_20_15_b.png', 
                       'p_14_12_b.png', 
                       'p_21_18_b.png', 
                       'p_18_16_b.png', 
                       'p_20_18_b.png', 
                       'p_10_9_b.png', 
                       'g_9_12_b.png', 
                       'g_12_16_b.png', 
                       'g_15_20_b.png', 
                       'g_12_14_b.png', 
                       'g_18_21_b.png', 
                       'g_16_18_b.png', 
                       'g_18_20_b.png', 
                       'g_9_10_b.png'
]

image_list_reverse = ['p_12_14_reverse.png', 
                      'p_12_16_reverse.png', 
                      'p_15_20_reverse.png', 
                      'p_16_18_reverse.png', 
                      'p_18_20_reverse.png', 
                      'p_18_21_reverse.png', 
                      'p_9_10_reverse.png', 
                      'p_9_12_reverse.png', 
                      'g_10_9_reverse.png', 
                      'g_12_9_reverse.png', 
                      'g_14_12_reverse.png', 
                      'g_16_12_reverse.png', 
                      'g_18_16_reverse.png', 
                      'g_20_15_reverse.png', 
                      'g_20_18_reverse.png', 
                      'g_21_18_reverse.png', 
                      'p_12_14_reverse_b.png', 
                      'p_12_16_reverse_b.png', 
                      'p_15_20_reverse_b.png', 
                      'p_16_18_reverse_b.png', 
                      'p_18_20_reverse_b.png', 
                      'p_18_21_reverse_b.png', 
                      'p_9_10_reverse_b.png', 
                      'p_9_12_reverse_b.png', 
                      'g_10_9_reverse_b.png', 
                      'g_12_9_reverse_b.png', 
                      'g_14_12_reverse_b.png', 
                      'g_16_12_reverse_b.png', 
                      'g_18_16_reverse_b.png', 
                      'g_20_15_reverse_b.png', 
                      'g_20_18_reverse_b.png', 
                      'g_21_18_reverse_b.png'
]

image_list_pink = ['l_p_12_9.png', 
                   'l_p_16_12.png', 
                   'l_p_20_15.png', 
                   'l_p_14_12.png', 
                   'l_p_21_18.png', 
                   'l_p_18_16.png', 
                   'l_p_20_18.png', 
                   'l_p_10_9.png', 
                   'r_p_9_12.png', 
                   'r_p_12_16.png', 
                   'r_p_15_20.png', 
                   'r_p_12_14.png', 
                   'r_p_18_21.png', 
                   'r_p_16_18.png', 
                   'r_p_18_20.png', 
                   'r_p_9_10.png', 
                   'l_p_12_9_a.png', 
                   'l_p_16_12_a.png', 
                   'l_p_20_15_a.png', 
                   'l_p_14_12_a.png', 
                   'l_p_21_18_a.png', 
                   'l_p_18_16_a.png', 
                   'l_p_20_18_a.png', 
                   'l_p_10_9_a.png', 
                   'r_p_9_12_a.png', 
                   'r_p_12_16_a.png', 
                   'r_p_15_20_a.png', 
                   'r_p_12_14_a.png', 
                   'r_p_18_21_a.png', 
                   'r_p_16_18_a.png', 
                   'r_p_18_20_a.png', 
                   'r_p_9_10_a.png'
]

image_list_green = ['l_g_12_9.png', 
                    'l_g_16_12.png', 
                    'l_g_20_15.png', 
                    'l_g_14_12.png', 
                    'l_g_21_18.png', 
                    'l_g_18_16.png', 
                    'l_g_20_18.png', 
                    'l_g_10_9.png', 
                    'r_g_9_12.png', 
                    'r_g_12_16.png', 
                    'r_g_15_20.png', 
                    'r_g_12_14.png', 
                    'r_g_18_21.png', 
                    'r_g_16_18.png', 
                    'r_g_18_20.png', 
                    'r_g_9_10.png', 
                    'l_g_12_9_a.png', 
                    'l_g_16_12_a.png', 
                    'l_g_20_15_a.png', 
                    'l_g_14_12_a.png', 
                    'l_g_21_18_a.png', 
                    'l_g_18_16_a.png', 
                    'l_g_20_18_a.png', 
                    'l_g_10_9_a.png', 
                    'r_g_9_12_a.png', 
                    'r_g_12_16_a.png', 
                    'r_g_15_20_a.png', 
                    'r_g_12_14_a.png', 
                    'r_g_18_21_a.png', 
                    'r_g_16_18_a.png', 
                    'r_g_18_20_a.png', 
                    'r_g_9_10_a.png'
]

image_sets = [
        (image_list_original, ['Pink', 'Green'], ['#FE9AFC', '#73ED51']),
        (image_list_reverse, ['Pink', 'Green'], ['#FE9AFC', '#73ED51']),
        (image_list_pink, ['Left', 'Right'], ['#FE9AFC', '#73ED51']),
        (image_list_green, ['Left', 'Right'], ['#FE9AFC', '#73ED51'])
    ]
 
results_dict = {
    'filename': [],
    'nL': [],
    'nR': [],
    'ratio': [],
    'correct_or_incorrect': [],
    'time_taken': [],
}

out1 = widgets.Output()

pink_colour = '#FE9AFC'  # MagentaPink
green_colour = '#73ED51'  # Green

button_labels = ['Pink', 'Green']
button_colours = [pink_colour, green_colour]
    
button_labels_reverse = ['Green', 'Pink']
button_colours_reverse = [green_colour, pink_colour]
    
button_labels_pink = ['Left', 'Right']
button_colours_pink = [pink_colour, pink_colour]
    
button_labels_green = ['Left', 'Right']
button_colours_green = [green_colour, green_colour]

event_info = {
    'type': '',
    'description': '',
    'time': -1
}

def wait_for_event(timeout=-1, interval=0.001, max_rate=20, allow_interupt=True):    
    start_wait = time.time()

    event_info['type'] = ""
    event_info['description'] = ""
    event_info['time'] = -1

    n_proc = int(max_rate*interval)+1
    
    with ui_events() as ui_poll:
        keep_looping = True
        while keep_looping==True:
            ui_poll(n_proc)

            if (timeout != -1) and (time.time() > start_wait + timeout):
                keep_looping = False
                
            if allow_interupt==True and event_info['description']!="":
                keep_looping = False
                
            time.sleep(interval)
            
    event_info['display_time'] = event_info['time'] - start_wait
    return event_info

def register_btn_event(btn):
    event_info['type'] = "button click"
    event_info['description'] = btn.description
    event_info['time'] = time.time()
    return

def run_set(set_type, button_labels, button_colours):

    if set_type == 'pink_set':
        image_list = image_list_pink
    elif set_type == 'green_set':
        image_list = image_list_green
    elif set_type == 'reverse_set':
        image_list = image_list_reverse
    else:
        set_type == 'original_set'
        image_list = image_list_original
    
    
    random.shuffle(image_list)
    
    for filename in image_list:
        image_data = Image(filename)
        image_data.width = 800  
        image_data.height = 400  
        #print(filename)
        match = re.search(r'\d+', filename)
        numbers = [int(match) for match in re.findall(r'\d+', filename)]
        nL = numbers[0]
        nR = numbers[1]
        ratio = min(numbers) / max(numbers)

        display(image_data)
        time.sleep(2)
        time_start = time.time()
        clear_output(wait=False)

        button_left = widgets.Button(description=button_labels[0], style={'button_color': button_colours[0]})
        button_right = widgets.Button(description=button_labels[1], style={'button_color': button_colours[1]})

        button_left.on_click(register_btn_event)
        button_right.on_click(register_btn_event)
        
        button_box = widgets.HBox([button_left, button_right])
        display(button_box)

        result = wait_for_event()

        time_end = time.time()
        time_taken = time_end - time_start

        correct_answer = image_data.filename[0].lower()
        
        if result['description'][0].lower() == correct_answer:
            correct_or_incorrect = 'correct'

        else:
            correct_or_incorrect = 'incorrect'
            
        clear_output(wait=True)

        results_dict['filename'].append(filename)
        results_dict['nR'].append(nR)
        results_dict['nL'].append(nL)
        results_dict['ratio'].append(ratio)
        results_dict['correct_or_incorrect'].append(correct_or_incorrect)
        results_dict['time_taken'].append(time_taken)

def ANS_test_3():
    data_consent_info = """DATA CONSENT INFORMATION:

    Please read:

    we wish to record your response data
    to an anonymised public data repository. 
    Your data will be used for educational teaching purposes
    practising data analysis and visualisation.

    Please type   yes   in the box below if you consent to the upload."""
    
    print(data_consent_info)
    result = input("> ") 
    
    if result == "yes": 
        print("Thanks for your participation.")
        print("Please contact philip.lewis@ucl.ac.uk")
        print("If you have any questions or concerns")
        print("regarding the stored results.")
        
    else: 
        raise(Exception("User did not consent to continue test."))
    
    id_instructions = """

    Enter your anonymised ID
    
    To generate an anonymous 4-letter unique user identifier please enter:
    - two letters based on the initials (first and last name) of a childhood friend
    - two letters based on the initials (first and last name) of a favourite actor / actress
    
    e.g. if your friend was called Charlie Brown and film star was Tom Cruise
         then your unique identifer would be CBTC
    """
    
    print(id_instructions)
    user_id = input("> ")
    
    print("User entered id:", user_id)

    print("Please enter your gender:")
    gender = input("")

    print("Welcome to the Approximate Number Sense test")
    time.sleep(1)
         
    print("The following images contain different numbers of pink and green dots. After being presented with each image, state whether there were more pink dots or green dots.")
    time.sleep(3)
    
    run_set('original_set', button_labels, button_colours)

    print("The green and pink dots have now reversed sides. After being presented with each image, state whether there were more green dots or pink dots.")
    time.sleep(3)
    
    run_set('reverse_set', button_labels_reverse, button_colours_reverse)

    print("The following images contain different numbers of pink dots only. After being presented with each image, state whether there were more dots on the left or right.")
    time.sleep(3)
    
    correct_answers_samecolours = [0]
    run_set('pink_set', button_labels_pink, button_colours_pink)

    print("The following images contain different numbers of green dots only. After being presented with each image, state whether there were more dots on the left or right.")
    time.sleep(3)
    run_set('green_set', button_labels_green, button_colours_green)

    print("End of test. Thank you for your participation.")

    results_df = pd.DataFrame(results_dict)
    #print(results_df)

    data_dict = {
        'user_id': user_id,
        'gender': gender,
        'results_json': results_df.to_json()
    }

    form_url = 'https://docs.google.com/forms/d/e/1FAIpQLSdvaMm76MB3v2TcSuAV47M9lDvI0EaRZagoVTCzgnEEiuMehw/viewform?usp=sf_link'
    success = send_to_google_form(data_dict, form_url)

    if success:
        print("Data successfully submitted to the Google Form.")
    else:
        print("Failed to submit data to the Google Form.")


ANS_test_3()
```
